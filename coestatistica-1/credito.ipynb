```python
aux = pd.DataFrame({'colunas' : credito.columns, 'tipos' : credito.dtypes}).reset_index().drop(columns = ['index'])
aux.
```

```python
aux_int64 = aux[aux['tipos'] == 'int64']
aux_int64
aux_int64[aux_int64['colunas'] != 'RowNumber'].reset_index(drop = True)
```

```python
pontuacao_estados = credito.groupby('estado_residencia').agg({'pontuacao_credito': [pd.Series.mode,np.median, np.mean, np.std]})

pontuacao_estados.columns = ['moda','mediana','media','desvio_padrao']

resultado = pontuacao_estados.reset_index().set_index('estado_residencia').transpose()

resultado
```

```python
resultado.to_json('submission.json')
```
